```python
import seaborn as sns
import pandas as pd
```

### kde plot 

```python
my_plot = sns.kdeplot(pokemon_data['Attack Point'])
my_plot.set_title('Attack point Distribution Plot')
my_plot.set_xlabel('Attack Point')
my_plot.set_ylabel('Frequency')
```

### displot 

```python
sns.displot(pokemon_data['Attack Point'])
```

### Rug Plot

```python
sns.rugplot(pokemon_data['Attack Point'],
           height = 0.5, axis = 'x')
```

### Bins

```python
my_plot = sns.distplot(pokemon_data['Attack Point'], kde = False, rug = True, bins = 100)
my_plot.set_title('Attack point Distribution Plot')
my_plot.set_xlabel('Attack Point')
my_plot.set_ylabel('Frequency')
```

```python
from matplotlib import pyplot as plt
```

```python
sns.set(color_codes = True)
sns.kdeplot(pokemon_data['Attack Point'],
            shade = True,
            color = 'g')
plt.show()
```

# Bivariate analysis

### Jointplot

```python
sns.jointplot (x = 'Attack Point', 
               y = 'Defense Point',
               data = pokemon_data)
plt.show()
```

### Playing with shapes

```python
sns.jointplot(x ='Attack Point',
              y ='Defense Point',
              data = pokemon_data,
              kind ='hex',
              color ='r')
```

```python
sns.jointplot(x ='Attack Point',
              y ='Defense Point',
              data = pokemon_data,
              kind ='kde',
              color ='b')
```

### cormat

```python
cormat = pokemon_data.corr()
plt.subplots(figsize = (7,7))
sns.heatmap(cormat, vmax = .8,
            square = True,
            annot = True, 
            fmt ='.2f', 
            cmap = 'Reds')
plt.show()
```

# Categorical Plot

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

### stripplot 

```python
sns.stripplot(x='day',
              y='tip',
              data = tips,
              )
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.stripplot(x='day',
              y='tip',
              data=tips,
              jitter = True
              )
plt.show()
```

### swarmplot

```python
plt.subplots(figsize=(15,5))
sns.swarmplot(x='day',
              y='tip',
              data=tips)
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.swarmplot(x='day',
              y='tip',
              data=tips, 
             hue = 'size')
plt.show()
```

### box plot

```python
plt.subplots(figsize=(15,5))
sns.boxplot(x='day',
            y='tip'
            ,data=tips)
plt.show()
```

### factorplot

```python
sns.factorplot(x='day',
               y='tip',
               data=tips,
               hue='size',
               size=5,
               aspect=2,
              kind='bar')
plt.show()
```
